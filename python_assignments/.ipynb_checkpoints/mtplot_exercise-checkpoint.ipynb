```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig1 = plt.figure(1)
x = np.linspace(0,10,500)
y = np.sin(x)
y1 = x**2
y2 = np.exp(-0.3*x+5)
y3 = np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.subplot(2,2,3)
plt.plot(x,y2)
plt.subplot(2,2,4)
plt.plot(x,y3)

```

```python
fig2=plt.figure(2)
x = np.linspace(0,10,500)
y = np.sin(x)
y1 = x**2
y2 = np.exp(-x+500)
y3 = np.cos(x)
sup1 = plt.subplot(2,1,1)
plt.plot(x,y)
sup1.set(xlim=(0,14),ylim=(-1,2))
sup1.xtick([])
plt.subplot(2,1,2)
plt.plot(x,y3)
```

```python
fig1.savefig('4plot', dpi=600)
fig2.savefig('2plot', dpi=200)
```
