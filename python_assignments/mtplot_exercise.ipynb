```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,500)
y = np.sin(x)
y1 = x**2
y2 = np.exp(-0.3*x+5)
y3 = np.cos(x)

fig1 = plt.figure(1)

plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.subplot(2,2,3)
plt.plot(x,y2)
plt.subplot(2,2,4)
plt.plot(x,y3)

```

```python
fig2=plt.figure(2)
#ax1 = fig2.add_axis( [left, bottom, width, height] )
sup1 = plt.subplot(2,1,1)
plt.plot(x,y)
sup1.set(xlim=(0,14),ylim=(-2,2), xticks=[], yticks=np.arange(-2,2.5,0.5), title='plot_manipulated', xlabel='x', ylabel= 'sin(x)')

sup2=plt.subplot(2,1,2)
plt.plot(x,y,x,y3)
sup2.legend(['sin','cos'])
sup2.set(xlabel='x')
```

```python
fig1.savefig('4plot', dpi=600)
fig2.savefig('2plot', dpi=200)
```
